```python
import pandas as pd
covid_df=pd.read_csv('italy-covid-daywise.csv')
covid_df
```

```python
covid_df.date
```

```python
covid_df['date']=pd.to_datetime(covid_df.date)
```

```python
covid_df['date']
```

```python
covid_df['year']=pd.DatetimeIndex(covid_df.date).year
covid_df['month']=pd.DatetimeIndex(covid_df.date).month
covid_df['day']=pd.DatetimeIndex(covid_df.date).day
covid_df['weekday']=pd.DatetimeIndex(covid_df.date).weekday
```

```python
covid_df
```

```python
covid_df_june=covid_df[covid_df.month==6]
covid_df_june
```

```python
covid_df_june=covid_df[covid_df.month==6][['new_cases','new_deaths','new_tests']]
covid_df_june
```

```python
covid_df_june_totals=covid_df[covid_df.month==6][['new_cases','new_deaths','new_tests']].sum()
covid_df_june_totals
```
