```python
import pandas as pd
import numpy as np
```

```python
crime=pd.read_csv('crime_data.csv')
crime.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
crime_scaled=scaler.fit_transform(crime.iloc[:,1:])
crime_scaled
```

```python
import scipy.cluster.hierarchy as sch
```

```python
Dendogram=sch.dendrogram(sch.linkage(crime_scaled, method='single'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac=AgglomerativeClustering(n_clusters=5,affinity='euclidean', linkage='single')
```

```python
crime_hc=ac.fit_predict(crime_scaled)
crime_hc
```

```python
crime.insert(5,'Clusters',crime_hc)
crime
```
