```python
from sklearn.ensemble import IsolationForest
import pandas as pd
```

```python
data=pd.read_csv("C:/Users/Administrator/Downloads/iris.csv",index_col=0)
```

```python
data_encoded=pd.get_dummies(data)
```

```python
data_encoded
```

```python
isof=IsolationForest(random_state=10,contamination=0.02)
isof.fit(data_encoded)
```

```python
y_outliers=isof.predict(data_encoded)
```

```python
y_outliers
```

```python
data_encoded['Scores']=isof.decision_function(data_encoded)
```

```python
data_encoded['Anomaly']=isof.predict(data_encoded.iloc[:,0:7])
```
