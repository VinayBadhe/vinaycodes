```python
import pandas as pd
```

```python
iris=pd.read_csv('iris (1).csv')
iris
```

```python
iris_encoded=pd.get_dummies(iris)
iris_encoded
```

```python
from sklearn.ensemble import IsolationForest
```

```python
isolationforest=IsolationForest(random_state=11, contamination=0.01)
isolationforest.fit(iris_encoded)
```

```python
iris_outliers=isolationforest.predict(iris_encoded)
iris_outliers
```

```python
iris_scores=isolationforest.decision_function(iris_encoded)
iris_scores
```

```python
iris['Anamoly']=iris_outliers
iris['Scores']=iris_scores
iris
```
