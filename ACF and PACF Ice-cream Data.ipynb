```python
import pandas as pd
import numpy as np
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
ice_cream_data=pd.read_csv('ice_cream.csv')
ice_cream_data.head()
```

```python
ice_cream_data.rename(columns={'DATE':'Date','IPN31152N':'Production'}, inplace=True)ice_cream_data
```

```python
ice_cream_data['Date']=pd.to_datetime(ice_cream_data.Date)
```

```python
ice_cream_data.set_index('Date',inplace=True)
```

```python
ice_cream_data
```

```python
plot_acf(ice_cream_data.Production,lags=100)
```

```python
plot_pacf(ice_cream_data.Production,lags=100)
```
