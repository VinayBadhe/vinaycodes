```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flights_df=sns.load_dataset('flights')
```

```python
flights_df
```

```python
flights1_df=flights_df.pivot("month","year","passengers")
flights1_df
```

```python
plt.title("No. of Passengers (1000s)")
sns.heatmap(flights1_df)
```

```python
plt.title("No. of Passengers (1000s)")
sns.heatmap(flights1_df, fmt="d", annot=True, cmap='Blues');
```
