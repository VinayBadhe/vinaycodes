```python
import pandas as pd
```

```python
movies=pd.read_csv('my_movies (1).csv')
movies
```

```python
movies=movies.iloc[:,5:]
movies
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
frequent_items=apriori(movies,min_support=0.1,use_colnames=True)
frequent_items
```

```python
rules=association_rules(frequent_items, metric='lift',min_threshold=2)
rules
```

```python
rules.sort_values('lift',ascending=False)[0:20]
```
