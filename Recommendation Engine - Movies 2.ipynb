```python
import pandas as pd
```

```python
movies_df=pd.read_csv('Movie (1).csv')
movies_df
```

```python
len(movies_df.userId.unique())
```

```python
len(movies_df.movie.unique())
```

```python
user_movie_df=movies_df.pivot(index='userId',columns='movie')
user_movie_df
```

```python
user_movie_df.index = movies_df.userId.unique()
user_movie_df
```

```python
user_movie_df=user_movie_df.fillna(0)
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
similarites_movies=1-pairwise_distances(user_movie_df.values,metric='cosine')
similarites_movies_df=pd.DataFrame(similarites_movies)
similarites_movies_df
```

```python
similarites_movies_df.index=movies_df.userId.unique()
similarites_movies_df.columns=movies_df.userId.unique()
similarites_movies_df.iloc[0:5, 0:5]
```
