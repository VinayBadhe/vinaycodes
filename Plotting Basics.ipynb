```python
import seaborn as sns
```

```python
a=sns.load_dataset('anscombe')
a.head()
```

```python
import matplotlib.pyplot as plt
```

```python
x=a['x']
```

```python
y=a['y']
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'o')
plt.suptitle("Basic Data")
```

```python
data_set_1 = a.iloc[0:10]
```

```python
data_set_2 = a.iloc[10:20]
```

```python
data_set_3 = a.iloc[20:30]
```

```python
data_set_4 = a.iloc[30:40]
```

```python
fig=plt.figure()
axes1=fig.add_subplot(2,2,1)
axes2=fig.add_subplot(2,2,2)
axes3=fig.add_subplot(2,2,3)
axes4=fig.add_subplot(2,2,4)
axes1.plot(data_set_1['x'],data_set_2['y'],'o')
axes2.plot(data_set_2['x'],data_set_2['y'],'o')
axes3.plot(data_set_3['x'],data_set_3['y'],'o')
axes4.plot(data_set_4['x'],data_set_4['y'],'o')
axes1.set_title('Data Set 1')
axes2.set_title('Data Set 2')
axes3.set_title('Data Set 3')
axes4.set_title('Data Set 4')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes3.set_xlabel('x')
axes3.set_ylabel('y')
axes4.set_xlabel('x')
axes4.set_ylabel('y')
fig.suptitle('ANSCOMBE DATA SET')
fig.set_tight_layout(True)
plt.show()
```
